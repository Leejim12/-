```python
import urllib.request as ur
from bs4 import BeautifulSoup as bs
import re 
import usecsv

soup = bs(ur.urlopen('http://quotes.toscrape.com/').read(), 'html.parser')

quote_res =[['명언','이름']]

quote = soup.find_all('div',{"class":"quote"})
for line in quote:
  q = re.findall(r"[A-Z].+[a-z]\.",line.text)
  by = re.findall(r"by [A-Z][A-Za-z \.]+",line.text)
  quote_res.append(q+by)

usecsv.writecsv('quote.csv', quote_res)

```

```python
import urllib.request as ur
from bs4 import BeautifulSoup as bs

soup = bs(ur.urlopen('https://news.daum.net/').read(), 'html.parser')
news = soup.find_all('a',{"class":"link_txt"})

for line in news:
  print(line.text[37:-33])
  print(line.get('href'))
```
